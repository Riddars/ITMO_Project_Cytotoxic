```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Укажите путь к вашему файлу CSV на Google Drive
file_path = 'https://raw.githubusercontent.com/Riddars/ITMO_Project_Cytotoxic/main/Data/1.Cytotoxicity%20(%D0%BF%D0%B5%D1%80%D0%B2%D1%8B%D0%B9).csv'

# Загрузите файл в DataFrame
DataFrame = pd.read_csv(file_path)
```

```python
# Предположим, у вас есть DataFrame df с интересующим столбцом 'название_столбца'
# Допустим, df - это ваш DataFrame

# Вычисляем количество уникальных значений и их долю
уникальные_значения = DataFrame['Cell type'].value_counts()
доля = (уникальные_значения / уникальные_значения.sum()) * 100

# Создаем горизонтальную столбчатую диаграмму
plt.figure(figsize=(8, 28))
bars = plt.barh(доля.index, доля)

# Настройки для диаграммы
plt.xlabel('Доля, %')
plt.ylabel('Уникальные значения')
plt.title('Горизонтальная диаграмма уникальных значений и их доли')

# Добавляем значения справа от столбцов
for bar, value in zip(bars, доля):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, f'{value:.1f}%', ha='left', va='center')

# Отображаем диаграмму
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(range(len(DataFrame['viability (%)'])), DataFrame['viability (%)'], color='blue', alpha=0.7)
plt.xlabel('Индексы строк')
plt.ylabel('Выживаемость (%)')
plt.title('Точечный график выживаемости')
plt.show()
```
