4 графика: время, концентрация, диаметр, потенциал

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузите данные из CSV-файла
data = pd.read_csv('https://raw.githubusercontent.com/Riddars/ITMO_Project_Cytotoxic/%D0%92%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/filtered_dataset_All_Decription.csv')

# Столбцы, для которых нужно создать графики
columns_to_plot = ['time (hr)', 'concentration (ug/ml)', 'Hydrodynamic diameter (nm)', 'Zeta potential (mV)']

# Создайте одну большую фигуру для все графики
plt.figure(figsize=(16, 6))

# Создайте гистограмму для каждого столбца
for i, column_name in enumerate(columns_to_plot):
    plt.subplot(1, 4, i+1)  # 1 строка, 3 столбца, i+1 - индекс текущего графика
    n, bins, patches = plt.hist(data[column_name], bins=40, color='skyblue', edgecolor='black', lw=0.5)
    for patch in patches:
        patch.set_linewidth(0.5)
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')

try:
    plt.savefig('C:\\MyProject\\ITMO_Project_Cytotoxic\\histogram_plot.png')
except Exception as e:
    print(f"Ошибка при сохранении файла: {e}")

# Отобразите все графики
plt.tight_layout()
plt.show()
```

График для типа теста

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузите данные из CSV-файла
data = pd.read_csv('https://raw.githubusercontent.com/Riddars/ITMO_Project_Cytotoxic/%D0%92%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/filtered_dataset_All_Decription.csv')

# Столбцы, для которых нужно создать графики
columns_to_plot = ['test']

# Создайте одну большую фигуру для все графики
plt.figure(figsize=(16, 6))

# Создайте гистограмму для каждого столбца
for i, column_name in enumerate(columns_to_plot):
    plt.subplot(1, 2, i+1)  # 1 строка, 2 столбца, i+1 - индекс текущего графика
    n, bins, patches = plt.hist(data[column_name], bins=40, color='skyblue', edgecolor='black', lw=0.5)
    for patch in patches:
        patch.set_linewidth(0.5)
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)  # Поворачиваем подписи по оси x на 90 градусов


try:
    plt.savefig('C:\\MyProject\\ITMO_Project_Cytotoxic\\histogram_plot.png')
except Exception as e:
    print(f"Ошибка при сохранении файла: {e}")

# Отобразите все графики
plt.tight_layout()
plt.show()
```
