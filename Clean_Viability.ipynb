```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Замените 'data' на вашу переменную с данными и 'column_name' на имя столбца
DataFrame = pd.read_csv('https://raw.githubusercontent.com/Riddars/ITMO_Project_Cytotoxic/main/Data/All_descriptors_dataset.csv')
column_name = 'viability (%)'

# Фильтрация данных
filtered_data = DataFrame[DataFrame['viability (%)'] <= 150]
```

```python
print(filtered_data.head())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(filtered_data['viability (%)'])), filtered_data['viability (%)'], color='blue', alpha=0.7)
plt.xlabel('Индексы строк')
plt.ylabel('Выживаемость (%)')
plt.title('Точечный график выживаемости после удаления выбросов')
plt.show()
```
