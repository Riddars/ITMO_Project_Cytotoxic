```python
DataFrame = pd.read_csv('https://raw.githubusercontent.com/Riddars/ITMO_Project_Cytotoxic/main/Cytotoxicity.csv', encoding='ISO-8859-1')
```

```python
DataFrame.head()
```

```python
import pubchempy as pcp
from pubchempy import Compound, get_compounds
# Химическая формула, для которой вы хотите получить CID
CID = []
for i in DataFrame["material"]:
  result = pcp.get_compounds(i, 'name')
  CID.append(result)
print(CID)
```

```python
DataFrame["CID"]=CID
```

```python
import re
# Замените все нецифровые символы в столбце на пустую строку
DataFrame['CID'] = DataFrame['CID'].apply(lambda x: re.sub(r'\D', '', str(x)))

# Преобразуйте столбец в числовой тип данных (int) и уберите десятичные знаки
DataFrame['CID'] = pd.to_numeric(DataFrame['CID'], errors='coerce').astype('Int64')

# Теперь ваш столбец содержит только целые числа без десятичных знаков и нулей в конце
```
